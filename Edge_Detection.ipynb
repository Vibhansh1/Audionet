```python
import math
import numpy as np
import matplotlib.pyplot as plt
import cv2

image_path = "/content/image.png"

image = cv2.imread(image_path)
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

gray = gray/255

mat = np.array(gray)

m, n = np.shape(mat)

Gx1 = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
Gx2 = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
Gy1 = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
Gy2 = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])

output = np.zeros([m-2, n-2])
output_x1 = np.zeros([m-2, n-2])
output_x2 = np.zeros([m-2, n-2])
output_y1 = np.zeros([m-2, n-2])
output_y2 = np.zeros([m-2, n-2])

for x in range(1, gray.shape[0]-2):
    for y in range(1, gray.shape[1]-2):
        output_x1[x, y] = (np.multiply(Gx1, gray[x-1:x+2, y-1:y+2])).sum()

for x in range(1, gray.shape[0]-2):
    for y in range(1, gray.shape[1]-2):
        output_x2[x, y] = (np.multiply(Gx2, gray[x-1:x+2, y-1:y+2])).sum()

for x in range(1, gray.shape[0]-2):
    for y in range(1, gray.shape[1]-2):
        output_y1[x, y] = (np.multiply(Gy1, gray[x-1:x+2, y-1:y+2])).sum()

for x in range(1, gray.shape[0]-2):
    for y in range(1, gray.shape[1]-2):
        output_y2[x, y] = (np.multiply(Gy2, gray[x-1:x+2, y-1:y+2])).sum()
output_v=abs(output_y1)+abs(output_y2)
output_h=abs(output_x1)+abs(output_x2)
output = abs(output_x1) + abs(output_x2) + abs(output_y1) + abs(output_y2)

for x in range(1, gray.shape[0]-2):
    for y in range(1, gray.shape[1]-2):
        if output[x, y]>0.5:
            output[x, y] = 255
        else:
            output[x, y] = 0

plt.imshow(output, cmap='gray')
plt.show()

plt.imshow(image)
plt.show()







```
